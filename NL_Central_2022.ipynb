```python
import numpy as np
import pandas as pd
import pybaseball as pyb
import matplotlib.pyplot as plt
```

```python
year = 2022

cardinals = pyb.schedule_and_record(year,'STL')
brewers = pyb.schedule_and_record(year,'MIL')
cubs = pyb.schedule_and_record(year,'CHC')
pirates = pyb.schedule_and_record(year,'PIT')
reds = pyb.schedule_and_record(year,'CIN')
```

# **Summary Statistics**

The Division winning Cardinals had a clear lead in Mean Runs. The Pirates are the only team with Mean Runs under 4. In Runs Allowed the Cardinals were the only team to have a mean under 4, while the last place Reds and Pirates had a Mean Runs allowed over 5. The Cardinals also had the hottest streak in the division at 8.

# **Wins, Win % & GB**

Gathering wins, losses, and win percentages for each team

```python
cardinals['Wins'] = np.where(cardinals['W/L']=='W',1,(np.where(cardinals['W/L']=='W-wo',
                                                               1,0))).cumsum()
cardinals['Losses'] = np.where(cardinals['W/L']=='L',1,(np.where(cardinals['W/L']=='L-wo',
                                                                 1,0))).cumsum()
cardinals['Win_Percentage'] = cardinals['Wins']/(cardinals['Wins'] + cardinals['Losses'])


brewers['Wins'] = np.where(brewers['W/L']=='W',1,(np.where(brewers['W/L']=='W-wo',
                                                          1,0))).cumsum()
brewers['Losses'] = np.where(brewers['W/L']=='L',1,(np.where(brewers['W/L']=='L-wo',
                                                             1,0))).cumsum()
brewers['Win_Percentage'] = brewers['Wins']/(brewers['Wins'] + brewers['Losses'])


cubs['Wins'] = np.where(cubs['W/L']=='W',1,(np.where(cubs['W/L']=='W-wo',
                                                    1,0))).cumsum()
cubs['Losses'] = np.where(cubs['W/L']=='L',1,(np.where(cubs['W/L']=='L-wo',
                                                       1,0))).cumsum()
cubs['Win_Percentage'] = cubs['Wins']/(cubs['Wins'] + cubs['Losses'])


pirates['Wins'] = np.where(pirates['W/L']=='W',1,(np.where(pirates['W/L']=='W-wo',
                                                           1,0))).cumsum()
pirates['Losses'] = np.where(pirates['W/L']=='L',1,(np.where(pirates['W/L']=='L-wo',
                                                             1,0))).cumsum()
pirates['Win_Percentage'] = pirates['Wins']/(pirates['Wins'] + pirates['Losses'])


reds['Wins'] = np.where(reds['W/L']=='W',1,(np.where(reds['W/L']=='W-wo',
                                                     1,0))).cumsum()
reds['Losses'] = np.where(reds['W/L']=='L',1,(np.where(reds['W/L']=='L-wo',
                                                       1,0))).cumsum()
reds['Win_Percentage'] = reds['Wins']/(reds['Wins'] + reds['Losses'])                                                                                                            
```

#**Graphing Wins & Win Percentages**

The NL Central was essentially a two team race between the Cardinals and Brewers. At about the 105 game mark the Cardinals took hold of the division lead and never relinquished. 

```python
plt.rcParams['figure.figsize']=(8,6)

plt.plot(cardinals['Wins'],label='Cardinals',c='lightcoral')
plt.plot(brewers['Wins'],label='Brewers',c='goldenrod')
plt.plot(cubs['Wins'],label='Cubs',c='blue')
plt.plot(pirates['Wins'],label='Pirates',c='black')
plt.plot(reds['Wins'],label='Reds',c='red')

plt.xticks(np.arange(0,len(cardinals.index),step=10))

plt.xlabel('Game')
plt.ylabel('Wins')
plt.legend(loc='lower right')
plt.grid(True)

plt.title("NL Central Wins Comparison({})".format(year))
```

```python
#Graphing Win Percentage

plt.rcParams['figure.figsize']=(8,6)

plt.xticks(np.arange(0,len(brewers.index),step=10))

plt.plot(brewers['Win_Percentage'],label='Brewers',c='darkgoldenrod')
plt.plot(cardinals['Win_Percentage'],label='Cardinals',c='lightcoral')
plt.plot(cubs['Win_Percentage'],label='Cubs',c='blue')
plt.plot(pirates['Win_Percentage'],label='Pirates',c='black')
plt.plot(reds['Win_Percentage'],label='Reds',c='red')

plt.xlabel('Game')
plt.ylabel('Wins')
plt.legend(loc='lower right')
plt.grid(True)

plt.title("NL Central Win Percentages ({})".format(year))
```

# **Graphing Games Back**

creating a modified column to numerically represent GB

```python
def modify_gb(gb):
  if gb == 'Tied':
    gb = 0.0
  elif 'up' in gb:
    gb = 0.0
  else:
    gb = -float(gb)

  return gb
```

```python
brewers['int_GB'] = brewers['GB'].dropna().apply(modify_gb)
cardinals['int_GB'] = cardinals['GB'].dropna().apply(modify_gb)
cubs['int_GB'] = cubs['GB'].dropna().apply(modify_gb)
pirates['int_GB'] = pirates['GB'].dropna().apply(modify_gb)
reds['int_GB'] = reds['GB'].dropna().apply(modify_gb)
```

# **Games Back Graph**

The Reds were the first team to fall 10 games back just before game 20. They never rebounded and stayed at a double digit deficit for the rest of the season. The 3rd place Cubs and 4th place Pirates never contended as well. Both teams along with the Reds stayed at least 10 games from first from game 70 to the end of the season. 

The Brewers fell a season high 9 games back at around game 130 after letting go of the division lead at around game 110.


```python
plt.rcParams['figure.figsize']=(8,6)

plt.plot(brewers['int_GB'],label='Brewers',c='darkgoldenrod')
plt.plot(cardinals['int_GB'],label='Cardinals',c='lightcoral')
plt.plot(cubs['int_GB'],label='Cubs',c='blue')
plt.plot(pirates['int_GB'],label='Pirates',c='black')
plt.plot(reds['int_GB'],label='Reds',c='red')

plt.xticks(np.arange(0,len(brewers.index),step=10))
plt.yticks(np.arange(0,-50, step=-5))
plt.xlabel('Game')
plt.ylabel('Games Back (in negative)')
plt.grid(True)

plt.legend(loc='lower left')
plt.title("NL Central Games Back Comparison ({})".format(year))
```

# **Graphing Total Runs & Runs Allowed**

Creating a new column for Total Runs and Total RA

```python
brewers['Total_Runs'] = brewers['R'].cumsum()
brewers['Total_RA'] = brewers['RA'].cumsum()

cardinals['Total_Runs'] = cardinals['R'].cumsum()
cardinals['Total_RA'] = cardinals['RA'].cumsum()

cubs['Total_Runs'] = cubs['R'].cumsum()
cubs['Total_RA'] = cubs['RA'].cumsum()

pirates['Total_Runs'] = pirates['R'].cumsum()
pirates['Total_RA'] = pirates['RA'].cumsum()

reds['Total_Runs'] = reds['R'].cumsum()
reds['Total_RA'] = reds['RA'].cumsum()
```

# **Total Runs/RA Graph**

The Reds scored runs at a similar rate as the cardinals and Brewers but, they were tied with the Pirates with the most runs allowed in the division. Both the Pirates and Reds were the only teams that gave up over 800 runs during the season. 

The Pirates not only gave up the most runs but they also scored the fewest runs scoring less than 600. 

The Cardinals took the lead in Runs Scored at about game 110 and this coincided with them taking over 1st place in the division as well. The Cardinals also took a brief lead around game 60 which was the same time they took the lead in Runs Allowed. Based on both graphs the Cardinals are the clear winner of the division.


```python
plt.rcParams["figure.figsize"]=(16,8)

plt.subplot(1,2,2)
plt.plot(brewers['Total_Runs'],label='Brewers',c='darkgoldenrod')
plt.plot(cardinals['Total_Runs'],label='Cardinals',c='lightcoral')
plt.plot(cubs['Total_Runs'],label='Cubs',c='blue')
plt.plot(pirates['Total_Runs'],label='Pirates',c='black')
plt.plot(reds['Total_Runs'],label='Reds',c='red')

plt.xlabel('Game')
plt.ylabel('Total Runs')
plt.legend(loc='lower right')
plt.grid(True)

plt.title("NL Central Total Runs Comparison ({})".format(year))


plt.rcParams["figure.figsize"]=(16,8)

plt.subplot(1,2,1)
plt.plot(brewers['Total_RA'],label='Brewers',c='darkgoldenrod')
plt.plot(cardinals['Total_RA'],label='Cardinals',c='lightcoral')
plt.plot(cubs['Total_RA'],label='Cubs',c='blue')
plt.plot(pirates['Total_RA'],label='Pirates',c='black')
plt.plot(reds['Total_RA'],label='Reds',c='red')

plt.xlabel('Game')
plt.ylabel('Runs Allowed')
plt.legend(loc='lower right')
plt.grid(True)
plt.title("NL Central Runs Allowed Comparison ({})".format(year))
```
